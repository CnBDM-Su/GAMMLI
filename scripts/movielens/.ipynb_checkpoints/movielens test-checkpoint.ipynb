## LVXNN

```python
import time
import numpy as np
import pandas as pd 
import tensorflow as tf
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from lvxnn.LVXNN import LV_XNN
from sklearn.metrics.pairwise import cosine_similarity
from gaminet.utils import local_visualize
from gaminet.utils import global_visualize_density
from gaminet.utils import feature_importance_visualize
from gaminet.utils import plot_trajectory
from gaminet.utils import plot_regularization
from sklearn.linear_model import LinearRegression
from lvxnn.DataReader import data_initialize
from collections import OrderedDict
```

```python
model_info['ui_shape']
```

```python
train= pd.read_csv('data/movie_lens/train.csv')
test = pd.read_csv('data/movie_lens/test.csv')

list1 = train.columns
meta_info = OrderedDict()
for i in list1:
    meta_info[i]={'type': 'categorical','source':'item'} 
meta_info['Occupation']={"type":"categorical",'source':'user'}
meta_info['Gender']={"type":"categorical",'source':'user'}
meta_info['Age'] = {"type":"continues",'source':'user'}
meta_info['user_id']={"type":"id",'source':'user'}
meta_info['item_id']={"type":"id",'source':'item'}
meta_info['target']={"type":"target",'source':''}


tr_x, tr_Xi, tr_y , te_x , te_Xi, te_y, meta_info, model_info = data_initialize(train,test,meta_info,"Regression")
```

```python
model = LV_XNN(model_info=model_info, meta_info=meta_info, subnet_arch=[8, 16],interact_arch=[20, 10],activation_func=tf.tanh, batch_size=1000, lr_bp=0.01, auto_tune=True,
               interaction_epochs=30,main_effect_epochs=30,tuning_epochs=30,loss_threshold_main=0.01,loss_threshold_inter=0.01,alpha=0.5,
              verbose=True,val_ratio=0.125, early_stop_thres=100,interact_num=10,u_group_num=10,i_group_num=20,scale_ratio=0.7,n_power_iterations=5,n_oversamples=0,
              mf_training_iters=150,mf_tuning_iters=100,change_mode=True,convergence_threshold=0.001,max_rank=3,shrinkage_value=20)
```

```python
st_time = time.time()
model.fit(tr_x,tr_Xi, tr_y)
ed_time = time.time()
```

```python
pred = model.predict(te_x, te_Xi)
mae = mean_absolute_error(te_y,pred)
rmse = mean_squared_error(te_y,pred)**0.5
print(mae,rmse)
```

```python
pred_v = model.final_gam_model.predict(te_x)
mae = mean_absolute_error(te_y,pred_v)
rmse = mean_squared_error(te_y,pred_v)**0.5
print(mae,rmse)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

train= pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
data,val = train_test_split(train,test_size=0.125)

x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
x_t = test.iloc[:,:-1].values
y_t = test.iloc[:,-1].values

enc = MinMaxScaler()
x = enc.fit_transform(x)
x_t = enc.fit_transform(x_t)

def auto_test(rank):
    mae = []
    rmse = []
    for times in range(10):
        xgb = XGBRegressor(n_jobs=-1)
        xgb.fit(x,y)
        pred = xgb.predict(x_t)
        
        mae.append(mean_absolute_error(y_t,pred))
        rmse.append(mean_squared_error(y_t,pred)**0.5)

    i_result = np.array([rank+1,np.mean(mae),np.mean(rmse)]).reshape(1,-1)
    result = pd.DataFrame(i_result,columns=['rank','mae','rmse'])
    
    return result

results = (auto_test(rank) for rank in range(1))
#results = Parallel(n_jobs=3)(delayed(auto_test)(rank) for rank in range(3))
results_ = pd.concat(results)
print(results_)
#results_.to_csv('sim_result_1.csv',index=None)
```
